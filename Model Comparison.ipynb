```python
import enso
import matplotlib.pyplot as plt
import numpy as np
import xarray
```

```python
T = 12
```

```python
datapath = '~/CM4_ctrl_indices.nc'
all_signals_nc = xarray.open_dataset(datapath, decode_times = False)
nino34 = all_signals_nc['nino34'].values
clm = enso.climatology(nino34)
anm = enso.anomaly(nino34, clm)
oni = enso.ONI(anm)
```

```python
(persistance_corr, persistance_rmse) = enso.persistance_corr_rmse(oni, T)
```

```python
print(persistance_rmse)
```

# All Signals

```python
linear_corr = np.loadtxt(fname = "linear_correlation")
linear_rmse = np.loadtxt(fname = "linear_RMSE")

#to optimize
simpleNN_corr = np.loadtxt(fname = "simpleNN_correlation")
simpleNN_rmse = np.loadtxt(fname = "simpleNN_RMSE") 

CNN_corr = np.loadtxt(fname = "CNN_correlation")
CNN_rmse = np.loadtxt(fname = "CNN_RMSE")

LSTM_corr = np.loadtxt(fname = "LSTM_correlation")
LSTM_rmse = np.loadtxt(fname = "LSTM_RMSE")
```

```python
plt.plot(persistance_corr, label = "Persistance");
plt.plot(linear_corr, label = "Linear NN");
plt.plot(simpleNN_corr, label = "Simple NN");
plt.plot(CNN_corr, label = "CNN");
plt.plot(LSTM_corr, label = "LSTM");
plt.ylabel('Correlation');
plt.xlabel('timeline');
plt.legend();
plt.figure();

plt.plot(persistance_rmse, label = "Persistance");
plt.plot(linear_rmse, label = "Linear NN");
plt.plot(simpleNN_rmse, label = "Simple NN");
plt.plot(CNN_rmse, label = "CNN");
plt.plot(LSTM_rmse, label = "LSTM");
plt.ylabel('RMSE');
plt.xlabel('timeline');
plt.legend();
```
